```python
df_obesity = pd.read_csv('../data/obesity.csv')
print("Shape:", df_obesity.shape)
print("Columns:", df_obesity.columns.tolist())
df_obesity.head(3)
```

```python
print("Sex values:", df_diabetes['Sex'].unique())
print("Age values:", df_diabetes['Age'].unique())
print("Year range:", df_diabetes['Year'].min(), "~", df_diabetes['Year'].max())
print("Number of countries:", df_diabetes['Country/Region/World'].nunique())
```

```python
df_diabetes = df_diabetes.rename(columns={
    'Country/Region/World': 'country',
    'ISO': 'iso',
    'Sex': 'sex',
    'Year': 'year',
    'Age': 'age',
    'Prevalence of diabetes (18+ years)': 'diabetes_prevalence',
    'Proportion of people with diabetes who were treated (30+ years)': 'treatment_rate'
})

df_obesity = df_obesity.rename(columns={
    'Entity': 'country',
    'Code': 'iso',
    'Year': 'year',
    'Prevalence of obesity among adults, BMI >= 30 (crude estimate) (%) - Sex: both sexes - Age group: 18+  years of age': 'obesity_prevalence'
})

print("Diabetes columns:", df_diabetes.columns.tolist())
print("Obesity columns:", df_obesity.columns.tolist())
```

```python
df_diabetes = df_diabetes.rename(columns={
    'Prevalence of diabetes (18+ years) lower 95% uncertainty interval': 'diabetes_lower_ci',
    'Prevalence of diabetes (18+ years) upper 95% uncertainty interval': 'diabetes_upper_ci',
    'Proportion of people with diabetes who were treated (30+ years) lower 95% uncertainty interval': 'treatment_lower_ci',
    'Proportion of people with diabetes who were treated (30+ years) upper 95% uncertainty interval': 'treatment_upper_ci'
})

print("All columns:", df_diabetes.columns.tolist())
```

```python
nordic = ['Sweden', 'Norway', 'Denmark', 'Finland', 'Iceland', 'Netherlands']

# Filter Nordic countries, both sexes
df_nordic = df_diabetes[
    df_diabetes['country'].isin(nordic)
].copy()

print("Nordic rows:", df_nordic.shape)
print("\nRows per country:")
print(df_nordic.groupby(['country', 'sex'])['year'].count())
print("\nYear range:", df_nordic['year'].min(), "~", df_nordic['year'].max())
```
